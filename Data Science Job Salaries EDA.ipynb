Examining data through Exploratory Data Analysis (EDA) is a critical aspect of any data-related project. It is essential to explore and comprehend the data and its attributes, as well as the inter-relationships between them to gain insights from the data. 

# Features/columns in the dataset

1. work_year: The year the salary was paid.
2. experience_level: The experience level in the job during the year.
3. employment_type: The type of employment for the role.
4. job_title: The role worked in during the year.
5. salary: The total gross salary amount paid.
6. salary_currency: The currency of the salary paid as an ISO 4217 currency code.
7. salaryinusd: The salary in USD.
8. employee_residence: Employee's primary country of residence in during the work year as an ISO 3166 country code.
9. remote_ratio: The overall amount of work done remotely
10. company_location: The country of the employer's main office or contracting branch as an ISO 3166 country code.
11. company_size: The median number of people that worked for the company during the year.

# 3.2. Top 3 Job Title by Experience Level

### What are the top 3 Job titles by experience level?


```python
#Top 3 Job Titles by Experience Level

# Group the data by experience level and job title, and count the occurrences
grouped = data.groupby(['job_title', 'experience_level']).size().reset_index(name='count')

# Sort the data by experience level and count in descending order
sorted_data = grouped.sort_values(['job_title', 'count'], ascending=[True, False])

# Get the top 3 job titles by experience level
top3 = sorted_data.groupby('job_title').head(3)

# Create a pivot table to reshape the data for plotting
pivot_table = top3.pivot(index='job_title', columns='experience_level', values='count')

# Plot the data as a bar chart (pivot table of aggregated data in order to make a stacked bar visualization)
pivot_table.plot(kind='bar', stacked=True)

# Add axis labels and experience_level title
plt.xlabel('job_title')
plt.ylabel('count')
plt.title('Top 3 Job Titles by Experience Level')

# Display the plot
plt.show()
```

OBSERVATION; The top 3 job titles by experience level are: Data Scientist, Data Engineer and Data Analyst

1. Data Scientist jobs offer more 'Entry-level'and'Junior Mid-level' positions compared to Data Engineer and Data Analyst jobs.
2. Data Engineer jobs offer more 'Senior-level'positions compared to Data Scientist and Data Analyst.
3. These top 3 job titles dont have 'Executive-level'positions.

# 5. Salary Analysis


## 5.1. Salary by work year


```python
#trends in salary over time

salary_trend=data.groupby('work_year')['salary_in_usd'].sum()
plt.figure(figsize=(10,5))
sns.lineplot(x=salary_trend.index, y=salary_trend.values, linewidth=3)

#xticks(marks the data points at the given positions with ticks )

xticks=[2020, 2021, 2022]
plt.xticks(xticks, xticks)
plt.xlabel('Year')
plt.ylabel('salary_in_usd')
plt.title('Salary Trend over the Years', size=10);
```

 OBSERVATION; we can see that 'salary' has continued to increase.The year with the greatest 'salary' earnings is 2022.

## 5.2. Salary by Experience Level


OBSERVATION; Employees in Mid Junior-Level position earn more Salary followed by 'Executive-Level', 'Entry-Level' and the least is 'Senior-Level'.

```python
#impact of salary on the level of experience

plt.figure(figsize=(10,5))
sns.barplot(x = data['experience_level'], y = data['salary_in_usd'])
plt.title('Salary Based on Experience level')
plt.xlabel('Experience Level')
plt.ylabel('Salary in usd')
plt.show()
```

OBSERVATION; Employees in Executive-level position(Experience level) earn more SALARY followed by 'Senior-Level', 'Mid Junior-Level' and the least is 'Entry-Level'.

## 5.3. Salary by Company Size

### Can salary be impacted by the size of a company?

```python
#Highest Salary Based on company Size

plt.figure(figsize=(10,5))
sns.barplot(x =data['company_size'] , y=data['salary_in_usd'])
plt.title('Highest Salary Based on Company Size')
plt.xlabel('company Size')
plt.ylabel('Salary')
plt.show()
```

OBSERVATION; from the graph we can see that Large companies pay higher salaries, followed by medium-sized ones. Most larger companies tend to have more resources and a remarkable ability to pay higher wages because larger companies usually have more significant revenue streams, larger budgets for employee compensation, and more competitive benefits packages.
